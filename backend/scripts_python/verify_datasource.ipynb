```python
import numpy as numpy
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
```

# Revisar columnas de la fuente de datos actual

```python
base_numeralia_actual  = pd.read_excel("C:/Users/ssocial10/Documents/proyectos/app-consulta-cartografica/backend/data/numeralia_fallas.xlsx")
base_numeralia_actual_gdf = gpd.GeoDataFrame(base_numeralia_actual, 
                                             geometry=[Point(xy) for xy in zip(base_numeralia_actual.LONGITUD, base_numeralia_actual.LATITUD)], 
                                             crs='EPSG:4326')

columnas_base_numeralia_actual_gdf = base_numeralia_actual_gdf.columns.to_list()
print(columnas_base_numeralia_actual_gdf)
```

```python
print(base_numeralia_actual_gdf.info())
print(base_numeralia_actual_gdf.shape)
```
